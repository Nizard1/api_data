```python
# OpenAlex endpoint to query for works
OPENALEX_API_WORKS = "https://api.openalex.org/works"

def query_personWork(orcid):
    page=1
    pages=1
    while page<=pages:
        params = {'filter': 'authorships.author.orcid:'+orcid, 'page': page}
        response = requests.get(
            url=OPENALEX_API_WORKS,
            params=params,
            headers= {'Accept': 'application/json'}
        )
        r=response.json()
        pages=r["meta"]["count"]//r["meta"]["per_page"]
    
        for i in range(r["meta"]["per_page"]):
            print(r["results"][i]["doi"],r["results"][i]["title"])
        page+=1
    
    
    
```
